```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('../Healthcare/dataset/heart.csv')
target = 'target'
```

## üß† 1. Understand the Target Variable

### Check the distribution of the outcome (e.g., target, heart_disease, diagnosis)
### ‚Üí How many people have heart disease vs. not?

### If it‚Äôs imbalanced (e.g., 90% ‚Äúno‚Äù and 10% ‚Äúyes‚Äù), models might be biased.

### Visualize with a simple bar chart to see the ratio of positive to negative cases.

```python

sns.countplot(x='target', data=df)    
plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

## ‚ù§Ô∏è 2. Analyze Key Risk Factors

### Look for medical, lifestyle, and demographic variables that might relate to heart disease:

```python
print(df.groupby(target)[['age']].mean())
```

### People with heart disease have a visibly higher median age, it confirms that age is a major demographic risk factor.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=target, y='age', data=df)
plt.title("Age Distribution by Heart Disease Status")
plt.show()
```

### Females (0) show higher counts in the positive class, it aligns with known higher female prevalence.

```python
sns.countplot(x='sex', hue=target, data=df)
plt.title("Heart Disease by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.show()
```
